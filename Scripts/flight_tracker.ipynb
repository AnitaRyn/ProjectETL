# Transform data

### Create a table of porting country

```python
port_country_db = df[["Port_Country"]].copy()
```

```python
# remove duplicates in country_port
port_country_db.drop_duplicates(inplace=True)
port_country_db.sort_values("Port_Country", inplace=True)

# allocate Id number to each country
port_country_db["id"] = port_country_db['Port_Country'].astype('category').cat.codes

# rearrange columns
port_country_db = port_country_db[["id", "Port_Country"]]

# set new index column
port_country_db.set_index("id",drop = True)
```

```python
port_country_db.to_csv("../Data Transformed/Port_Country.csv", index = False)
```

### Data cleaning airline name columns - Testing

```python
air_line_df = df[["Airline"]].drop_duplicates()
air_line_df.reset_index()
```

```python
air_line_df2 = df2[["Name"]].drop_duplicates()
```

```python
air_line_df2
```

```python
# merge trial to see if the airline name matched
matching_df = air_line_df.merge(air_line_df2, left_on="Airline", right_on="Name",how = "left")
```

```python
# matching result after merging
print(matching_df.loc[matching_df["Name"].isnull()])
print("----------")
print("Unmatched airlines:")
print(matching_df.loc[matching_df["Name"].isnull()].count())
```

```python
# Replace common words within airline name such as "Airlines", "Intl", "International"...for having better matching name:
# e.g : Canadian Airlines Intl vs Canadian Airlines;
# Air Caledonie International  vs Air Caledonie;
# Olympic Airways vs Olympic Airlines;
# Nauru Air Corporation vs Air Nauru.

# Use title() to avoid all upper case e.g: EVA Air vs Eva Air 
# remove multiple spaces within name after the above replacements

# Define function to clean column names
def clean(column):
    column = column.str.strip().str.replace("Airlines","", case = False).str.strip().str.title()
    column = column.str.strip().str.replace("Airways","", case = False).str.strip().str.title()
    column = column.str.strip().str.replace("A/l","", case = False).str.strip().str.title()
    column = column.str.strip().str.replace("Intl","", case = False).str.strip().str.title()
    column = column.str.strip().str.replace("International","", case = False).str.strip().str.title()
    column = column.str.strip().str.replace("Corporation","", case = False).str.strip().str.title()
    column = column.str.strip().str.replace("Air","", case = False).str.strip().str.title()
    column = column.replace('\s+', ' ', regex=True)
    return column

air_line_df["Airline"] = clean(air_line_df["Airline"])

air_line_df2["Name"] = clean(air_line_df2["Name"])
```

```python
matching_df = air_line_df.merge(air_line_df2, left_on="Airline", right_on="Name",how = "left")
```

```python
print(matching_df.loc[matching_df["Name"].isnull()])
print("----------")
print("Unmatched airlines:")
print(matching_df.loc[matching_df["Name"].isnull()].count())
```
