```python
# Import dependencies
import pandas as pd
import numpy as np
```

### Extract CSVs into DataFrames

```python
airlines_kaggle_file = "../Data Extracted/airlines.csv"
airlines_df = pd.read_csv(airlines_kaggle_file) 
airlines_df.head()
```

```python
# Look for missing values/ if any
airlines_df.count()
```

### Rename "Name" column header to "Airline

```python
clean_df = airlines_df = airlines_df.rename(columns={'Name':'Airline'})
clean_df
```

### Count Duplicates in the "Airline" column

```python
name_unique_count = airlines_df['Airline'].nunique()
name_unique_count
```

### Drop Duplicates in "Airline" Column

```python
clean_df = airlines_df.drop_duplicates(subset=['Airline'])
clean_df
```

```python
clean_df = clean_df.drop([0,0])
```

```python
clean_df.to_csv("../Data Transformed/airlines_cleaned.csv", index = False)
```
